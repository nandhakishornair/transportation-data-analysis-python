```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Explanation for x = x.reshape(-1,1):
# This line reshapes the NumPy array 'x'.
# The .reshape() method changes the shape of an array without changing its data.
# -1 in the first dimension tells NumPy to automatically calculate the number of rows
# based on the total number of elements in the array and the specified number of columns.
# 1 in the second dimension means the array will have one column.
# For example, if 'x' was a 1D array of 50 elements, it would become a 2D array of shape (50, 1).
# This reshaping is often necessary when preparing 1D data for machine learning models
# that expect 2D input (e.g., each row representing a sample and each column a feature).
```

```python
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 2 * x + 3 + np.random.normal(0, 2, 50)  # linear pattern with noise
x = x.reshape(-1,1)
```

```python
linear = LinearRegression()
linear.fit(x, y)
y_linear = linear.predict(x)
```

```python
y_constant = np.full_like(y, y.mean())  # predicts mean only
```

```python
poly = PolynomialFeatures(degree=10)
x_poly = poly.fit_transform(x)
overfit = LinearRegression()
overfit.fit(x_poly, y)
y_overfit = overfit.predict(x_poly)
```

```python
plt.scatter(x, y, color='black', label='Data')
plt.plot(x, y_linear, color='blue', label='Linear fit (good)')
plt.plot(x, y_constant, color='green', label='Underfit')
plt.plot(x, y_overfit, color='red', label='Overfit')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Underfitting vs Good Fit vs Overfitting")
plt.show()
```
