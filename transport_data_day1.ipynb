```python
import pandas as pd
```

```python
data = {
    "ROUTE":["X1","X2","X1","X2"],
    "STOP":["A","B" ,"B" ,"A" ],
    "TRAVEL_TIME_MIN":[12,13,22,11]
}
```

```python
#average travel time per route
df = pd.DataFrame(data)
```

```python
avg_travel_time = df.groupby("STOP")["TRAVEL_TIME_MIN"].mean()
```

```python
print("average avg_travel_time per route in min::",avg_travel_time)
```
