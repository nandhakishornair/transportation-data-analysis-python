```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv"
df = pd.read_csv(url)
df = df.dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["state"].head(10),df["total exports"].head(10))
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.title("top 10 routes (states)by total exports")
plt.xlabel("route (State)")
plt.ylabel("Total passenger Exports")
plt.show()
```

seaborn quick start

```python
# correlation between multiple columns
plt.figure(figsize=(10, 6))
sns.heatmap(df[['beef','pork','poultry','total exports']].corr(),annot=True, cmap = "Reds")
plt.title("correlation Matrix - Transport indicators")
plt.show()
```
