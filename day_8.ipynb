###we check for erroneous data and fill/remove them if necessary

```python
import pandas as pd
import numpy as np
# example data with missing /wrong values
```

```python
data ={
    "route": ["A", "B", "C", "D", "E"],
     "Travel_Time_min": [30, 45, np.nan, 55, 40],
    "Speed_kmph": [60, np.nan, 50, -10, 70]  # -10 is invalid
}
```

```python
df = pd.DataFrame(data)
print("oringnal data is :")
print(df)
```

```python
#finding missing values
print("\nMissing values in the data:")
print(df.isnull().sum())
```

```python
# fill missing values with mean values
df["Travel_Time_min"].fillna(df["Travel_Time_min"].mean(), inplace=True)
df['Speed_kmph'].fillna(df['Speed_kmph'].mean(), inplace=True)
```

```python
#replace the invalid data with the mean value
df['Travel_Time_min'].fillna(df['Travel_Time_min'].mean(), inplace=True)
df['Speed_kmph'].fillna(df['Speed_kmph'].mean(), inplace=True)
#fill negatieve val with mean
df.loc[df["Speed_kmph"] < 0, "Speed_kmph"] = np.nan
df["Speed_kmph"].fillna(df["Speed_kmph"].mean(), inplace=True)
# 4️⃣ Round values
df["Travel_Time_min"] = df["Travel_Time_min"].round(1)
df["Speed_kmph"] = df["Speed_kmph"].round(1)

print("\nCleaned data:")
print(df)
```
