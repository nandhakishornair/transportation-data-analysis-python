```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)
```

```python
#make 100 trips
distance = np.random.randint(2, 40, 100)
avg_speed = np.random.randint(20, 75, 100)
stops = np.random.randint(0, 10, 100)
```

```python
#travel time formula with randomness
time = (distance / avg_speed) * 60 + stops * 1.5 + np.random.normal(0, 2, 100)
```

```python
df = pd.DataFrame(
    {'distance': distance,
     'avg_speed': avg_speed,
     'stops': stops,
     'time': time})
```

```python
df.head()
```

```python
#creation of scatter plots
plt.scatter (df['distance'], df['time'])
plt.xlabel('Distance')
plt.ylabel('Time')
plt.title('Scatter Plot of Distance vs. Time')
sns.pairplot(df)
plt.show()
```

```python
corr = df.corr()
print(corr)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
X = df[["distance", "avg_speed", "stops"]]
y = df["time"]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python

print("R-squared:", r2_score(y, y_pred))
```
