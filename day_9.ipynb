```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#create sample data
data = pd.DataFrame({ "Route": ["A", "B", "C", "D", "E", "F"],
    "Speed_kmph": [60, 70, 45, 80, 55, 90],
    "Travel_Time_min": [50, 40, 60, 35, 55, 30],
    "Distance_km": [50, 55, 45, 50, 50, 45],
    "Fuel_Used_L":  [5, 6, 7, 4, 5.5, 6.5]})
```

```python
#scatter plot - speed vs travel time
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Speed_kmph",
                y="Travel_Time_min",
                hue="Route",
                palette="Set2",
                s = 100,
                data=data)
plt.title("Speed vs Travel Time", fontsize=16)
plt.xlabel("Speed (km/h)", fontsize=14)
plt.ylabel("Travel Time (min)", fontsize=14)
plt.legend(title="Route", fontsize=12)
plt.grid(True)
plt.show()
```

```python
#correlation analysis
corr = data[["Speed_kmph", "Travel_Time_min","Distance_km","Fuel_Used_L"]].corr()
print("Correlation Matrix:")
print(corr)
```

```python
#optional heat map visulization
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#creating scatter plot between fuel used and distance
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Distance_km",
                y="Fuel_Used_L",
                hue="Route",
                palette="coolwarm",
                s = 100,
                data=data)
```
