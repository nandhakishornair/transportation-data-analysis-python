```python
import pandas as pd
import numpy as np

#create a messy dataset
data = {'Name': ['John', 'Jane', 'Mike', 'Emily', 'David'],
        'Age': [25,np.nan,30,35,np.nan],
        'Salary': [50000,60000,70000,np.nan,80000]}
df = pd.DataFrame(data)
print("original dataframe")
print(df)
```

```python
#find missing values
print("missing values")
print(df.isnull().sum())
```

```python
#remove row with missing valuese
df_clean = df.dropna()
print("cleaned dataframe after dropping missing values")
print(df_clean)
```

```python
#filling missing values
df_filled = df.fillna({
    'Age': df['Age'].mean(),
    'Salary': df['Salary'].median()
})
print("dataframe after filling missing values")
print(df_filled)
```

```python
#remove duplicate rows
df_nodup = df_filled.drop_duplicates()
print("dataframe after removing duplicates")
print(df_nodup)
```
