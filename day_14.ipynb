```python
import pandas as pd
import numpy as np

data =  {'area_sqft': [1000, 1500, 2000, np.nan, 1800, 1800],
    'bedrooms': [2, 3, 4, 3, np.nan, 3],
    'price': [200000, 300000, 400000, 350000, 360000, 360000]
}
df = pd.DataFrame(data)
df.to_csv('housing_data.csv', index=False)
print("csv created")
```

```python
df= pd.read_csv('housing_data.csv')
print("first 5 row")
print(df.head())
print("Summary")
print(df.info())
print("\n Missing value")
print(df.isnull().sum())
print("\n Duplicate value")
print(df.duplicated().sum())
```

```python
df['area_sqft'] = df['area_sqft'].fillna(df['area_sqft'].mean())
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['area_sqft'], df['price'])
plt.xlabel('Area (sqft)')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```

```python
plt.hist(df['bedrooms'])
plt.xlabel('Bedrooms')
plt.ylabel('Frequency')
plt.title('Distribution of Bedrooms')
plt.show()
```

```python
plt.bar(df['area_sqft'], df['price'])
plt.xlabel('Area (sqft)')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```
