```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#time in hours
time = [0, 1, 2, 3, 4, 5]
# Speeds (km/h) on three routes
speed_A = [0, 20, 45, 50, 40, 30]
speed_B = [0, 15, 35, 45, 38, 25]
speed_C = [0, 10, 30, 40, 32, 22]
speed_D = [0, 25, 50, 48, 35, 28]

#plot each line
plt.figure(figsize=(7,5))
plt.plot(time, speed_A, marker='o', label='Route A', color='red')
plt.plot(time, speed_B, marker='s', label='Route B', color='blue')
plt.plot(time, speed_C, marker='*', label='Route C', color='orange')
plt.plot(time, speed_D, marker='^', label='Route C', color='green')
# Labels and title
plt.title("Vehicle Speeds Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Speed (km/h)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Routes")
plt.show()
```
