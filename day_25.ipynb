```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create example data
np.random.seed(42)
distance = np.linspace(1, 50, 100)
fuel_used = distance * 0.6 + np.random.normal(0, 3, 100)

df = pd.DataFrame({"distance": distance, "fuel": fuel_used})

# Train model
X = df[['distance']]
y = df['fuel']

model = LinearRegression()
model.fit(X, y)

# Predictions
predicted = model.predict(X)

# Residuals
residuals = y - predicted

# Plot residuals
plt.scatter(df['distance'], residuals)
plt.axhline(0, color='red')
plt.xlabel("Distance")
plt.ylabel("Residual (Actual - Predicted)")
plt.title("Residual Plot")
plt.show()

# Print sample values
print("First 5 residuals:")
print(residuals.head())
```
