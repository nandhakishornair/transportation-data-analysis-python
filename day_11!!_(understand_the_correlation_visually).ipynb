```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#set style
sns.set_style('darkgrid')
```

```python
#create data for different correlation levels
np.random.seed(101) #if u dont set seed the randomness happen, the random seed makes the random number fixed
x = np.linspace(0,10,100)
```

```python
#strong positieve correlatiion
y1 = x * 0.8 + np.random.normal(0, 5, 100)

# Weak positive correlation
y2 = x * 0.3 + np.random.normal(0, 20, 100)

# Strong negative correlation
y3 = -x * 0.8 + np.random.normal(0, 5, 100)

# Weak negative correlation
y4 = -x * 0.3 + np.random.normal(0, 20, 100)
```

```python
# Put all in one dataframe
df = pd.DataFrame({
    'x': np.tile(x, 4),
    'y': np.concatenate([y1, y2, y3, y4]),
    'type': ['Strong +ve']*100 + ['Weak +ve']*100 + ['Strong -ve']*100 + ['Weak -ve']*100
})

# Plot 4 panels
sns.lmplot(x='x', y='y', data=df, col='type', hue='type',
           col_wrap=3, ci=None, palette='coolwarm', scatter_kws={'s':20})
plt.suptitle('Different Types of Correlations', y=1.02)
plt.show()
```
