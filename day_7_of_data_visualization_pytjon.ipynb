 seaborn line visuals

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data  = pd.DataFrame({'time': [0,1,2,3,4,5]*3,
                      'speed':[0,20,45,50,40,30,  0,15,35,45,38,25,  0,10,30,40,32,22],
                      "Route": ["A"]*6 + ["B"]*6 + ["C"]*6})
```

plot

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="time", y="speed", hue="Route", data=data)
plt.title("Line Plot")

plt.show()
```
