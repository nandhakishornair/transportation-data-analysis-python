```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from statistics import mean
import seaborn as sns

# Example vehicle speed data
time_a = [0,1,2,3,4,5]
speed_a = [0,20,45,34,23,45]

time_b = [0,1,2,3,4,5]
speed_b = [0,15,35,30,25,40]

time_c = [0,1,2,3,4,5]
speed_c = [0,10,25,28,20,30]

# Averages
avg_timeA = mean(time_a)
avg_timeB = mean(time_b)
avg_timeC = mean(time_c)

avg_speedA = mean(speed_a)
avg_speedB = mean(speed_b)
avg_speedC = mean(speed_c)

# Create dataframe
data = pd.DataFrame({
    "Route": ["A", "B", "C"],
    "Average Time (h)": [avg_timeA, avg_timeB, avg_timeC],
    "Average Speed (km/h)": [avg_speedA, avg_speedB, avg_speedC]
})

# Seaborn bar plot
plt.figure(figsize=(6,4))
sns.barplot(x="Route", y="Average Speed (km/h)", data=data, palette="coolwarm")

plt.title("Average Vehicle Speed on Different Routes")
plt.xlabel("Route")
plt.ylabel("Average Speed (km/h)")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```
