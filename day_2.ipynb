```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 1: Create fake transport data
np.random.seed(42)
distance = np.linspace(1, 50, 100)  # km
fuel_used = distance * 0.6 + np.random.normal(0, 3, 100)  # liters

df = pd.DataFrame({"distance": distance, "fuel": fuel_used})

# Step 2: Create model
model = LinearRegression()

X = df[['distance']]   # independent variable (must be 2D)
y = df['fuel']         # dependent variable

model.fit(X, y)

# Step 3: Get predictions
pred = model.predict(X)

# Step 4: Plot
plt.scatter(distance, fuel_used)
plt.plot(distance, pred)
plt.title("Fuel vs Distance - Regression Line")
plt.xlabel("Distance (km)")
plt.ylabel("Fuel Used (L)")
plt.show()

# Step 5: Print results
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("R²:", model.score(X, y))
```
