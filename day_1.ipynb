```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([1,2,3,4]).reshape(-1,1)
y = np.array([3,5,7,9])

model = LinearRegression()
model.fit(x, y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of the original data points
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x.flatten(), y=y, s=100, color='blue', label='Data Points')

# Plot the regression line
# Predict y values using the trained model
y_pred = model.predict(x)
sns.lineplot(x=x.flatten(), y=y_pred, color='red', linewidth=2, label='Regression Line')

plt.title('Linear Regression Model Visualization')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
